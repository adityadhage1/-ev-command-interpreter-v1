```python
# Mumbai Green Gap Tracker - Day 4 Project
def calculate_savings(km, traffic_level):
    # Emissions in grams per km (India 2026 Avg)
    PETROL_BASE = 125
    EV_BASE = 80 # Lifecycle avg including grid

    # Mumbai Traffic Idling Penalty (Petrol wastes more in jams)
    traffic_multipliers = {"low": 1.0, "medium": 1.3, "high": 1.6}
    penalty = traffic_multipliers.get(traffic_level.lower(), 1.0)

    petrol_total = km * PETROL_BASE * penalty
    ev_total = km * EV_BASE
    savings = petrol_total - ev_total

    print(f"\n--- Mumbai Trip Report ({km}km, {traffic_level} traffic) ---")
    print(f"Petrol Emission: {petrol_total/1000:.2f} kg CO2")
    print(f"EV Emission:     {ev_total/1000:.2f} kg CO2")
    print(f"Green Savings:   {savings/1000:.2f} kg CO2 Saved! 🍀")

dist = float(input("Enter distance in km: "))
traffic = input("Traffic level (low/medium/high): ")
calculate_savings(dist, traffic)
```
